```python
m = df0.dropna()
i = df1.dropna()
r = df2.dropna()
h = df3.dropna()
n = df4.dropna()
```

```python
m['Starting Date'] = pd.to_datetime(n['Starting Date'])
i['Starting Date'] = pd.to_datetime(n['Starting Date'])
r['Starting Date'] = pd.to_datetime(n['Starting Date'])
h['Starting Date'] = pd.to_datetime(n['Starting Date'])
n['Starting Date'] = pd.to_datetime(n['Starting Date'])

m['Starting_Year'] = m['Starting Date'].dt.year
i['Starting_Year'] = i['Starting Date'].dt.year
r['Starting_Year'] = r['Starting Date'].dt.year
h['Starting_Year'] = h['Starting Date'].dt.year
n['Starting_Year'] = n['Starting Date'].dt.year

mm = m[m['Starting_Year'] == 2019]
ii = i[i['Starting_Year'] == 2019]
rr = r[r['Starting_Year'] == 2019]
hh = h[h['Starting_Year'] == 2019]
nn = n[n['Starting_Year'] == 2019]
```

```python
mmr = mm.drop(['Project Code', 'Title', 'Starting Date', 'Status', 'Source of Financing', 'Sovereign'], axis = 1)
iib = ii.drop(['Project Code', 'Title', 'Starting Date', 'Status', 'Source of Financing', 'Sovereign'], axis = 1)
rrt = rr.drop(['Project Code', 'Title', 'Starting Date', 'Status', 'Source of Financing', 'Sovereign'], axis = 1)
hhu = hh.drop(['Project Code', 'Title', 'Starting Date', 'Status', 'Source of Financing', 'Sovereign'], axis = 1)
nnj = nn.drop(['Project Code', 'Title', 'Starting Date', 'Status', 'Source of Financing', 'Sovereign'], axis = 1)
data = [mmr, iib, rrt, hhu, nnj]
df = pd.concat(data)

df['Commitment in U.A'] = df['Commitment in U.A'].str.replace(r',', '') #error raise because of , when we try to make this column as numeric
df['Commitment in U.A'] = df['Commitment in U.A'].apply(pd.to_numeric)
df.info()
df.to_csv('mydata.csv')
```

```python
df0_mmri = df.query('Sector == "Agriculture and Rural Development"')
df0_iib = df.query('Sector == "Finance"')
df0_rrt = df.query('Sector == "Transport"')
df0_hhu = df.query('Sector == "Multi-Sector"')
df0_nnj = df.query('Sector == "Power"')
```

```python
mq = df0_hhu['Commitment in U.A'].sum()
iq = df0_rrt['Commitment in U.A'].sum()
rq = df0_nnj['Commitment in U.A'].sum()
hq = df0_mmri['Commitment in U.A'].sum()
nq = df0_iib['Commitment in U.A'].sum()

yf0 = df0_iib.query('Country == "Multinational"')
yf1 = df0_iib.query('Country == "Morocco"')
yf2 = df0_iib.query('Country == "Kenya"')
multinational_percent = (yf0['Commitment in U.A'].sum() / nq) * 100
morocco1_percent = (yf1['Commitment in U.A'].sum() / nq) * 100
kenya1_percent = (yf2['Commitment in U.A'].sum() / nq) * 100

ya0 = df0_mmri.query('Country == "Côte d\'Ivoire"')
ya1 = df0_mmri.query('Country == "Congo"')
ya2 = df0_mmri.query('Country == "Sudan"')
cote_percent = (ya1['Commitment in U.A'].sum() / hq) * 100
congo_percent = (ya1['Commitment in U.A'].sum() / hq) * 100
sudan_percent = (ya2['Commitment in U.A'].sum() / hq) * 100

yp0 = df0_nnj.query('Country == "Angola"')
yp1 = df0_nnj.query('Country == "Mozambique"')
yp2 = df0_nnj.query('Country == "Morocco"')
angola_percent = (yp0['Commitment in U.A'].sum() / rq) * 100
mozambique_percent = (yp1['Commitment in U.A'].sum() / rq) * 100
morocco2_percent = (yp2['Commitment in U.A'].sum() / rq) * 100

yt0 = df0_rrt.query('Country == "Uganda"')
yt1 = df0_rrt.query('Country == "Kenya"')
yt2 = df0_rrt.query('Country == "Tanzania, United Republic of"')
uganda_percent = (yt0['Commitment in U.A'].sum() / iq) * 100
kenya2_percent = (yt1['Commitment in U.A'].sum() / iq) * 100
tanzania_percent = (yt2['Commitment in U.A'].sum() / iq) * 100

ym0 = df0_hhu.query('Country == "Niger"')
ym1 = df0_hhu.query('Country == "Rwanda"')
ym2 = df0_hhu.query('Country == "Madagascar"')
niger_percent = (ym0['Commitment in U.A'].sum() / mq) * 100
rwanda_percent = (ym1['Commitment in U.A'].sum() / mq) * 100
madagascar_percent = (ym2['Commitment in U.A'].sum() / mq) * 100
```

```python
rest_finance = 100 - (multinational_percent + morocco1_percent + kenya1_percent)
rest_agriculture = 100 - (cote_percent + congo_percent + sudan_percent)
rest_power = 100 - (angola_percent + mozambique_percent + morocco2_percent)
rest_transport = 100 - (uganda_percent + kenya2_percent + tanzania_percent)
rest_multi = 100 - (niger_percent + rwanda_percent + madagascar_percent)
```

```python
dfchart0 = pd.DataFrame({'finance sector':[multinational_percent, morocco1_percent, kenya1_percent, rest_finance]}, index = ['Multinational', 'Morocco', 'Kenya', 'Rest of Africa'])
dfchart1 = pd.DataFrame({'agriculture sector': [cote_percent, congo_percent, sudan_percent, rest_agriculture]}, index = ['Cote d\'Ivoire', 'Congo', 'Sudan', 'Rest of Africa'])
dfchart2 = pd.DataFrame({'power sector': [angola_percent, mozambique_percent, morocco2_percent, rest_power]}, index = ['Angola', 'Mozambique', 'Morocco', 'Rest of Africa'])
dfchart3 = pd.DataFrame({'transport sector': [uganda_percent, kenya2_percent, tanzania_percent, rest_transport]}, index = ['Uganda', ' Kneya', 'Tanzania', 'Rest of Africa'])
dfchart4 = pd.DataFrame({'multi sector': [niger_percent, rwanda_percent, madagascar_percent, rest_multi]}, index = ['Niger', 'Rwanda', 'Madagascar', 'Rest of Africa'])
```

```python
dfchart0.plot.pie(y = 'finance sector', figsize = (8,8))
```
