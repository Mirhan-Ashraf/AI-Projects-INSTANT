```python
colms = ["Id","ActivityDate","TotalSteps","VeryActiveDistance","FairlyActiveMinutes","LightlyActiveMinutes","SedentaryMinutes","Calories" ]
df = df[colms]
```

```python
df["TotalMin"] = df.VeryActiveDistance + df.FairlyActiveMinutes + df.SedentaryMinutes + df.LightlyActiveMinutes
```

```python
df["TotalHours"] = round(df.TotalMin/60)
```

```python
plt.figure(figsize= ((6,4)))
plt.hist(df.DayOfWeeek, bins=7, color= "blue", width = 0.4)
plt.grid(True)

plt.xlabel("Day of Week")
plt.ylabel("Frequency")

plt.title("Frequency by day of week")
plt.show()
```

```python
plt.figure(figsize=((8,6)))
plt.scatter(df.TotalSteps , df.Calories , c= df.Calories)

median_steps = 7405
median_calories = 2134

plt.axhline(median_calories , color = "blue" , label = "median of calories")
plt.axvline(median_steps , color = "red" , label = "median of steps")


plt.xlabel("steps")
plt.ylabel("calories")

plt.title("Calories By Total Steps")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=((8,6)))
plt.scatter(df.TotalHours , df.Calories , c= df.Calories)

median_hours = 23
median_calories = 2134

plt.axhline(median_calories , color = "blue" , label = "median of calories")
plt.axvline(median_hours , color = "red" , label = "median of hours")


plt.xlabel("hours")
plt.ylabel("calories")

plt.title("Calories By Total Hours")
plt.legend()
plt.show()
```
