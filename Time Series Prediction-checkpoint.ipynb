# Model

```python
train_set = data1.loc[data1.index < '01-01-2014']
test_set= data1.loc[data1.index >= '01-01-2014']
```

```python
fig,ax = plt.subplots(figsize = (16,6))
train_set.plot(ax=ax)
test_set.plot(ax=ax)
ax.axvline('01-01-2014', color= 'Black', ls='--')

plt.legend(['Train_Set','Test_set'])
plt.title('Train and Test Split')
plt.show()
```

```python
x_train = train_set.drop('PJME_MW',axis=1)
y_train = train_set['PJME_MW']

x_test = test_set.drop('PJME_MW',axis=1)
y_test = test_set['PJME_MW']
```

```python
model1 = XGBRegressor(n_estimator=1000, early_stopping_round=50)
```

```python
model1.fit(x_train, y_train)
```

```python
model1.score(x_train, y_train)
```

```python
premodel = model1.predict(x_test )
```

```python
mean_squared_error(premodel,y_test)
```

```python
np.sqrt(mean_squared_error(premodel,y_test))
```
